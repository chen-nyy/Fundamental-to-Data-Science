**Exercise 26:** Write a function called exponent(base, exp) that returns an int value of base raises to the power of exp.
Note here exp is a non-negative integer, and the base is an integer.

**Expected Output:**

Case 1:

base = 2
exponent = 5

2 raises to the power of 5 is: 32 i.e. (2 *2 * 2 *2 *2 = 32)

Case 2:

base = 5
exponent = 4

5 raises to the power of 4 is: 625 i.e. (5 *5 * 5 *5 = 625)

``````python
## calculate exponent function
def exponent(base, exp):
	if exp < 0:
		print("Exp should be non-negative integer")
		return 1

	result = base ** exp

	return result

## Case 1
print("Case 1:")
base = 2
exp = 5
print("base =", base, "exp =", exp)
## call function
result = exponent(base, exp)

ie = str(base)
for i in range(exp - 1):
	ie += "*" + str(base)

ie += "=" + str(result)
## result display
if exp >= 0:
	print(base, " raises to the power of ", exp, " is: ", result, " i.e. (", ie, ")", sep="")


## Case 2
print("Case 2:")
base = 5
exp = 4
print("base =", base, "exp =", exp)
## call function
result = exponent(base, exp)

ie = str(base)
for i in range(exp - 1):
	ie += "*" + str(base)

ie += "=" + str(result)
## result display
if exp >= 0:
	print(base, " raises to the power of ", exp, " is: ", result, " i.e. (", ie, ")", sep="")
``````
