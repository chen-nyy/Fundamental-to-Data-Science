```python
# Exercise 15
# Calculate covariance of two numerical variable and explain the result 

num_var2 = "Likes"
covariance = df[num_var].cov(df[num_var2])
print("Covariance:", covariance)
```

```python
# Exercise 16
# Calculate Correlation of two numerical variable and explain the result 
# What is the difference between correlation and covariance?

correlation = df[num_var].corr(df[num_var2])
print("Correlation:", correlation)
```

### What is the difference between correlation and covariance?
Covariance:
- shows how two variables change together
- can be any value
- depends on units
- hard to understand
Correlation:
- shows strength and direction
- unitless
- easier to compare across datasets
- easy to understand
